### ‚Ññ1. –°—Ç–µ–ø–µ–Ω–∏ –¥–≤–æ–π–∫–∏.

–¢–µ–ø–µ—Ä—å –≤—ã–≤–µ–¥–∏—Ç–µ –≤—Å–µ —Å—Ç–µ–ø–µ–Ω–∏ –¥–≤–æ–π–∫–∏ –≤ –¥–∏–∞–ø–∞–∑–æ–Ω–µ –æ—Ç [l, r]* –≤–∫–ª—é—á–∏—Ç–µ–ª—å–Ω–æ

*r - —Ç–µ–ø–µ—Ä—å –º–æ–∂–µ—Ç –ø—Ä–∏–Ω–∏–º–∞—Ç—å –±–æ–ª—å—à–∏–µ –∑–Ω–∞—á–µ–Ω–∏—è. –û–ø—Ç–∏–º–∏–∑–∏—Ä—É–π—Ç–µ —Å–≤–æ–π –∫–æ–¥.

–í—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ:

$1<=l<=r<=10^{10}$

–í—ã–≤–µ–¥–∏—Ç–µ –≤—Å–µ —á–∏—Å–ª–∞ –≤ –æ–¥–Ω—É —Å—Ç—Ä–æ–∫—É —á–µ—Ä–µ–∑ –ø—Ä–æ–±–µ–ª.



```bash
9 512
```

```bash
16
32
64
128
256
512
```


````python
l, r = map(int, input().split())
num = 1

while num <= r:
  if num >= l and num <= r:
    print(num)
  num = num * 2
````

### ‚Ññ2. –î–µ–ª–∏—Ç–µ–ª–∏ —á–∏—Å–ª–∞.
–í—ã–≤–µ–¥–∏—Ç–µ –≤—Å–µ –¥–µ–ª–∏—Ç–µ–ª–∏ —á–∏—Å–ª–∞ –æ—Ç 1 –¥–æ $n$ –≤–∫–ª—é—á–∏—Ç–µ–ª—å–Ω–æ.

–í—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ:
$1<=n<=10^7$

*–ë—É–¥—å—Ç–µ –∞–∫–∫—É—Ä–∞—Ç–Ω—ã —Å –∫–≤–∞–¥—Ä–∞—Ç–∞–º–∏

```bash

```
```bash

```

````python
n = int(input())

count = 1

for i in range(n):
  if n % count == 0:
    print(count)

  count += 1
````

### ‚Ññ3. –í–∏—Å–æ–∫–æ—Å–Ω—ã–π –≥–æ–¥.

–ü–æ –Ω–æ–º–µ—Ä—É –≥–æ–¥–∞ –æ–ø—Ä–µ–¥–µ–ª–∏—Ç–µ, —è–≤–ª—è–µ—Ç—Å—è –ª–∏ –¥–∞–Ω–Ω—ã–π –≥–æ–¥ –≤–∏—Å–æ–∫–æ—Å–Ω—ã–º.

–ù–∞–ø–æ–º–Ω–∏–º, —á—Ç–æ –≤–∏—Å–æ–∫–æ—Å–Ω—ã–º–∏ —è–≤–ª—è—é—Ç—Å—è –≥–æ–¥–∞, –Ω–æ–º–µ—Ä–∞ –∫–æ—Ç–æ—Ä—ã—Ö –∫—Ä–∞—Ç–Ω—ã 4, –Ω–æ –Ω–µ –∫—Ä–∞—Ç–Ω—ã 100, –∞ —Ç–∞–∫–∂–µ –≥–æ–¥–∞, –Ω–æ–º–µ—Ä–∞ –∫–æ—Ç–æ—Ä—ã—Ö –∫—Ä–∞—Ç–Ω—ã 400

–í—ã–≤–µ–¥–∏—Ç–µ `–î–∞` –∏–ª–∏ `–ù–µ—Ç` –≤ –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–∏ –æ—Ç –æ—Ç–≤–µ—Ç–∞ –Ω–∞ –≤–æ–ø—Ä–æ—Å –∑–∞–¥–∞—á–∏.

–í—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ:
```bash
2007
```

–†–µ–∑—É–ª—å—Ç–∞—Ç:
```bash
–ù–µ—Ç
```

````python
year = int(input())

if year % 4 == 0 and year % 100 != 0:
  print('–î–∞')
elif year % 400 == 0:
  print('–î–∞')
else:
  print('–ù–µ—Ç')
````

### ‚Ññ4. –®–∞—Ö–º–∞—Ç—ã.

–ü–æ–ª–µ —à–∞—Ö–º–∞—Ç–Ω–æ–π –¥–æ—Å–∫–∏ –æ–ø—Ä–µ–¥–µ–ª—è–µ—Ç—Å—è –ø–∞—Ä–æ–π –Ω–∞—Ç—É—Ä–∞–ª—å–Ω—ã—Ö —á–∏—Å–µ–ª, –∫–∞–∂–¥–æ–µ –∏–∑ –∫–æ—Ç–æ—Ä—ã—Ö –Ω–µ –ø—Ä–µ–≤–æ—Å—Ö–æ–¥–∏—Ç 8.

–ü–æ –≤–≤–µ–¥–µ–Ω–Ω—ã–º –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç–∞–º –¥–≤—É—Ö –ø–æ–ª–µ–π (ùëò,ùëô) –∏ (ùëö,ùëõ) –≤—ã—è—Å–Ω–∏—Ç–µ, —É–≥—Ä–æ–∂–∞–µ—Ç –ª–∏ —Ñ–µ—Ä–∑—å, –Ω–∞—Ö–æ–¥—è—â–∏–π—Å—è –Ω–∞ –ø–æ–ª–µ (ùëò,ùëô), –ø–æ–ª—é (ùëö,ùëõ)?

–í—ã–≤–µ–¥–∏—Ç–µ `–î–∞` –∏–ª–∏ `–ù–µ—Ç` –≤ –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–∏ –æ—Ç –æ—Ç–≤–µ—Ç–∞ –Ω–∞ –≤–æ–ø—Ä–æ—Å –∑–∞–¥–∞—á–∏.

–í—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ:
```bash
1 1 2 2
```

–†–µ–∑—É–ª—å—Ç–∞—Ç:
```bash
–î–∞
```

````python
k, l, m, n = map(int, input().split())

if abs(m - k) == abs(n - l):
  print('–î–∞')
else:
  print('–ù–µ—Ç')
````

### ‚Ññ5 k-–∞—è —Å–µ–∫—É–Ω–¥–∞ —Å—É—Ç–æ–∫

### ‚Ññ6 –ï–ª–æ—á–∫–∞

–í—ã–≤–µ–¥–∏—Ç–µ –µ–ª–æ—á–∫—É –≤—ã—Å–æ—Ç—ã $n$ –ø–æ —à–∞–±–ª–æ–Ω—É –∏–∑ –ø—Ä–∏–º–µ—Ä–∞

–í—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ:
```bash
3
```

–†–µ–∑—É–ª—å—Ç–∞—Ç:
```bash
  *
 ***
*****
```

````python
n = int(input())
count = n
star_count = 1

for i in range(n):
  print(' ' * count + '*' * star_count)
  count -= 1
  star_count += 2
````

### ‚Ññ7 –ú–∞–∫—Å–∏–º—É–º –∏ –º–∏–Ω–∏–º—É–º

–î–∞–Ω–∞ –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç—å, —Å–æ—Å—Ç–æ—è—â–∞—è –∏–∑ $n$ —á–∏—Å–µ–ª. –ù–∞–π—Ç–∏ –≤ –Ω–µ–π –º–∞–∫—Å–∏–º—É–º –∏ –º–∏–Ω–∏–º—É–º.

*–ó–∞–ø—Ä–µ—â–µ–Ω–æ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏–µ —Ñ—É–Ω–∫—Ü–∏–π min –∏ max


–í—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ - —á–∏—Å–ª–æ $n > 0$ –∏ —á–∏—Å–ª–∞ –Ω–∞ —Ä–∞–∑–Ω—ã—Ö —Å—Ç—Ä–æ—á–∫–∞—Ö $-1000 <= a_i <= 1000$:
```bash
3
5
6
7
```

–í—ã—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ:
```bash
5 7
```

````python
n = int(input())
max = int(input())
min = max

for i in range(n - 1):
  x = int(input())

  if x > max:
    max = x
  if x < min:
    min = x

print(min, max)
````

### ‚Ññ8 –ú–∏–Ω–∏–º—É–º –∏ –º–∏–Ω–∏–º—É–º

–î–∞–Ω–∞ –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç—å, —Å–æ—Å—Ç–æ—è—â–∞—è –∏–∑ $n$ —á–∏—Å–µ–ª. –ù–∞–π—Ç–∏ –≤ –Ω–µ–π –¥–≤–∞ —Å–∞–º—ã—Ö –º–∞–ª–µ–Ω—å–∫–∏—Ö —á–∏—Å–ª–∞.

*–ó–∞–ø—Ä–µ—â–µ–Ω–æ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏–µ —Ñ—É–Ω–∫—Ü–∏–π min –∏ max

–í—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ - —á–∏—Å–ª–æ $n > 0$ –∏ —á–∏—Å–ª–∞ –Ω–∞ —Ä–∞–∑–Ω—ã—Ö —Å—Ç—Ä–æ—á–∫–∞—Ö $-1000 <= a_i <= 1000$:
```bash
3
5
6
7
```

–í—ã—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ:
```bash
5 6
```

````python
n = int(input())

min1 = int(input())
min2 = int(input())

if min2 < min1:
  min1, min2 = min2, min1

for i in range(n - 2):
  x = int(input())

  if x < min1:
    min2 = min1
    min1 = x

print(min1, min2)
````
