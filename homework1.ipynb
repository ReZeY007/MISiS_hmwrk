### №1. Степени двойки.

Теперь выведите все степени двойки в диапазоне от [l, r]* включительно

*r - теперь может принимать большие значения. Оптимизируйте свой код.

Входные данные:

$1<=l<=r<=10^{10}$

Выведите все числа в одну строку через пробел.



```bash
9 512
```

```bash
16
32
64
128
256
512
```


````python
l, r = map(int, input().split())
num = 1

while num <= r:
  if num >= l and num <= r:
    print(num)
  num = num * 2
````

### №2. Делители числа.
Выведите все делители числа от 1 до $n$ включительно.

Входные данные:
$1<=n<=10^7$

*Будьте аккуратны с квадратами

```bash
21
```
```bash
1 3 7 21
```

````python
n = int(input())

count = 1

for i in range(n):
  if n % count == 0:
    print(count)

  count += 1
````

### №3. Високосный год.

По номеру года определите, является ли данный год високосным.

Напомним, что високосными являются года, номера которых кратны 4, но не кратны 100, а также года, номера которых кратны 400

Выведите `Да` или `Нет` в зависимости от ответа на вопрос задачи.

Входные данные:
```bash
2007
```

Результат:
```bash
Нет
```

````python
year = int(input())

if year % 4 == 0 and year % 100 != 0:
  print('Да')
elif year % 400 == 0:
  print('Да')
else:
  print('Нет')
````

### №4. Шахматы.

Поле шахматной доски определяется парой натуральных чисел, каждое из которых не превосходит 8.

По введенным координатам двух полей (𝑘,𝑙) и (𝑚,𝑛) выясните, угрожает ли ферзь, находящийся на поле (𝑘,𝑙), полю (𝑚,𝑛)?

Выведите `Да` или `Нет` в зависимости от ответа на вопрос задачи.

Входные данные:
```bash
1 1 2 2
```

Результат:
```bash
Да
```

````python
k, l, m, n = map(int, input().split())

if abs(m - k) == abs(n - l):
  print('Да')
else:
  print('Нет')
````

### №5 k-ая секунда суток

Идёт 𝑘-я секунда суток. Определите, сколько целых часов ℎ и целых минут m прошло с начала суток. Например, если

$𝑘=13257=3⋅3600+40⋅60+57$

Входные данные:
```bash
13257
```

Результат:
```bash
3 hours 40 minutes
```

````python
k = int(input())
hours = k // 3600
minutes = (k % 3600) // 60

print(f'{hours} hours {minutes} minutes')
````

### №6 Елочка

Выведите елочку высоты $n$ по шаблону из примера

Входные данные:
```bash
3
```

Результат:
```bash
  *
 ***
*****
```

````python
n = int(input())
count = n
star_count = 1

for i in range(n):
  print(' ' * count + '*' * star_count)
  count -= 1
  star_count += 2
````

### №7 Максимум и минимум

Дана последовательность, состоящая из $n$ чисел. Найти в ней максимум и минимум.

*Запрещено использование функций min и max


Входные данные - число $n > 0$ и числа на разных строчках $-1000 <= a_i <= 1000$:
```bash
3
5
6
7
```

Выходные данные:
```bash
5 7
```

````python
n = int(input())
max = int(input())
min = max

for i in range(n - 1):
  x = int(input())

  if x > max:
    max = x
  if x < min:
    min = x

print(min, max)
````

### №8 Минимум и минимум

Дана последовательность, состоящая из $n$ чисел. Найти в ней два самых маленьких числа.

*Запрещено использование функций min и max

Входные данные - число $n > 0$ и числа на разных строчках $-1000 <= a_i <= 1000$:
```bash
3
5
6
7
```

Выходные данные:
```bash
5 6
```

````python
n = int(input())

min1 = int(input())
min2 = int(input())

if min2 < min1:
  min1, min2 = min2, min1

for i in range(n - 2):
  x = int(input())

  if x < min1:
    min2 = min1
    min1 = x

print(min1, min2)
````
